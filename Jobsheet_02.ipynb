# 1. Hasil Perkalian

```python
import numpy as np

# Definisi Matriks
A = np.array([[3, 1], [2, 0]])
B = np.array([[2, 1], [1, -1]])
C = np.array([[-1, 1, 2], [2, 0, -1]])

# a. A x B (Perkalian Matriks)
hasil_AB = np.matmul(A, B)
print("Hasil A x B:\n", hasil_AB)

# b. B . C
# Catatan: Secara matematis, perkalian matriks (dot product) mensyaratkan
# jumlah kolom matriks pertama sama dengan jumlah baris matriks kedua.
# Karena B (2x2) dan C (3x2), maka operasi B . C memerlukan transpose pada C (C.T)
# agar menjadi (2x2) * (2x3).
hasil_BC = np.dot(B, C)
print("Hasil B . C (Transpose):\n", hasil_BC)
```

# 3. Analisis Vektor 3D

```python
import numpy as np

# Definisi Vektor
A = np.array([5, 3, 7])
B = np.array([12, -3, 1])

# Menentukan panjang (norm) masing-masing vektor
panjang_A = np.linalg.norm(A)
panjang_B = np.linalg.norm(B)

# Hasil perkalian titik (Dot Product) A.B
dot_product = np.dot(A, B)

# Hasil perkalian silang (Cross Product) AxB
cross_product = np.cross(A, B)

print(f"Panjang Vektor A: {panjang_A}")
print(f"Panjang Vektor B: {panjang_B}")
print(f"Hasil A.B (Dot Product): {dot_product}")
print(f"Hasil AxB (Cross Product): {cross_product}")
```
