# 1. Central Tendency

```python
# a. Mendapatkan Nilai Mean
import numpy as np
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.mean(score)

# b. Mendapatkan Nilai Median
import numpy as np
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.median(score)

# c. Mendapatkan Nilai Modus / Mode
from scipy import stats
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = stats.mode(score, keepdims=True)
```

# 2. Spread

```python
# a. Mendapatkan Nilai Range
import numpy as np
data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]
minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum

# b. Mendapatkan Nilai Percentile
import numpy as np
data = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.percentile(data, 90)

# c. Mendapatkan Nilai Quartile dan Interquartile Range
import numpy as np
data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]
Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1
Q1, Q3, IQR

# d. Mendapatkan Nilai Variance dan Standart Deviation
import numpy as np
data= [86, 87, 88, 86, 87, 85, 86]
mean = np.mean(data)
var = np.var(data)
std = np.std(data)
mean, var, std
```

# 1. Mencari Modus Data Pelemparan Dadu

```python
from scipy import stats

# Input data pelemparan dadu
dadu = [1, 2, 3, 4, 5, 5, 6, 2, 3, 4, 5, 6, 6, 4, 3, 2, 1, 4, 3, 5, 6, 6, 5, 4, 5]

# Menghitung modus
hasil_modus = stats.mode(dadu, keepdims=True)

print(f"Modus dari data dadu adalah: {hasil_modus.mode[0]}")
print(f"Frekuensi kemunculan: {hasil_modus.count[0]}")
```

# 2. Menentukan Persentil ke-25 dan ke-75

```python
import numpy as np

# Input deret data
deret_data = [9, 10, 11, 6, 8, 7, 7, 5, 4, 5]

# Menghitung persentil ke-25 dan ke-75
p25 = np.percentile(deret_data, 25)
p75 = np.percentile(deret_data, 75)

print(f"Persentil ke-25: {p25}")
print(f"Persentil ke-75: {p75}")
```

# 3. Analisis Data Hasil Ujian (Quartile, IQR, dan Outlier)

```python
import numpy as np

# Input data hasil ujian
nilai_ujian = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

# A. Urutkan data
nilai_urut = sorted(nilai_ujian)
print(f"Data Terurut: {nilai_urut}")

# B. Identifikasi Q1, Q3, dan IQR
Q1 = np.percentile(nilai_urut, 25)
Q3 = np.percentile(nilai_urut, 75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")

# C. Tentukan Lower Bound dan Upper Bound
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# D. Identifikasi Outlier
outliers = [x for x in nilai_urut if x < lower_bound or x > upper_bound]
print(f"Data Outlier: {outliers}")
```

# 4. Menghitung Standar Deviasi Data Berkelompok

```python
import numpy as np

# Menggunakan nilai tengah (midpoint) untuk setiap rentang kelas
# (41+45)/2 = 43, dst.
midpoints = np.array([43, 48, 53, 58, 63, 68])
frequencies = np.array([2, 3, 6, 4, 5, 2])

# Menghitung Rata-rata (Mean) untuk data berkelompok
mean_kelompok = sum(midpoints * frequencies) / sum(frequencies)

# Menghitung Varians
squared_diff = (midpoints - mean_kelompok) ** 2
variance_kelompok = sum(frequencies * squared_diff) / (sum(frequencies) - 1)

# Menghitung Standar Deviasi [cite: 289]
std_kelompok = np.sqrt(variance_kelompok)

print(f"Rata-rata Berat Badan: {mean_kelompok:.2f} kg")
print(f"Standar Deviasi: {std_kelompok:.2f}")
```
